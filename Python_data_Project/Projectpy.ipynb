```python
#Importing data,colums,pandas and Plots
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('Uk_jobs.XLSX')

print(df.info()) 

print(df.columns)
```

```python
#Cleaning data

df.dropna(inplace= True)
```

```python
#functions for creating graphes 
def large_ind(location):
    
    location = location.sort_values(by=['Jobs 2014'])
    graph = location[['Industry','Jobs 2011','Jobs 2014']].tail(5)
    graph.plot.barh(x='Industry', y=['Jobs 2011','Jobs 2014'],color=['orange','purple'])
    plt.show()
```

```python
def small_ind(location):
    
    location = location.sort_values(by=['Jobs 2014'])
    graph = location[['Industry','Jobs 2011','Jobs 2014']].head(5)
    graph.plot.barh(x='Industry', y=['Jobs 2011','Jobs 2014'],color=['red','blue'])
    plt.show()
```

```python
def grow_industry(location):
    
    location = location.sort_values(by=['% Change'])
    grab=location[['Industry','% Change']].tail(5)
    grab.plot.bar(x='Industry',y='% Change',color=['green'])
    plt.show()
```

```python
def decline_industry(location):
    location = location.sort_values(by=['% Change'])
    grab=location[['Industry','% Change']].head(5)
    grab.plot.bar(x='Industry',y='% Change',color=['brown'])
    plt.show()
```

```python
#Extracting london 

data = pd.read_excel('Uk_jobs.XLSX',index_col = "City")

lon=data.loc[["London"]]

large_ind(lon)

small_ind(lon)

grow_industry(lon)

decline_industry(lon)

```
